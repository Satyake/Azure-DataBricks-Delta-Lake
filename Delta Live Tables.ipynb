```
import dlt
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```

orders_path='/mnt/streaming-demo/full_dataset/orders_full.csv'
schema="""
ORDER_ID INT,
ORDER_DATETIME STRING,
CUSTOMER_ID INT,
ORDER_STATUS STRING,
STORE_ID INT

"""


order_items_path='/mnt/streaming-demo/full_dataset/order_items.csv'
schema_items="""
ORDER_ID INT,
LINE_ITEM_ID INT,
PRODUCT_ID INT,
UNIT_PRICE DOUBLE,
QUANTITY INT

"""
```

```

@dlt.table
def order_items_bronze():
    return spark.read.format('csv').option('header','True').schema(schema_items).load(order_items_path)

```

```

@dlt.table
def order_bronze():
    return spark.read.format('csv').option('header','True').schema(schema).load(orders_path)
```

```
@dlt.table
def orders_silver():
    return (
        dlt.read('order_bronze').select(
            'order_id',\
                to_date('order_datetime','dd-MMM-yy kk.mm.ss.SS').alias('ORDER_DATE'),\
                    'CUSTOMER_ID', \
                    'ORDER_STATUS', \
                    'STORE_ID',
                    current_timestamp().alias('MODIFIED_DATE')
        )
    )
```

```
@dlt.table
def order_items_silver():
    return dlt.read('order_items_bronze').select('order_id','product_id','unit_price','quantity',current_timestamp().alias('Modified_Date'))
```
